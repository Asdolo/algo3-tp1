```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
pandas_dataframe_fb = pd.read_csv('output/fuerzabruta_out.csv')
pandas_dataframe_bt = pd.read_csv('output/backtracking_out.csv')
pandas_dataframe_pd = pd.read_csv('output/programaciondinamica_out.csv')
```

```python
df = pandas_dataframe_fb

df['tiempo (backtracking)'] = pandas_dataframe_bt['tiempo (backtracking)']
df['tiempo (programaciondinamica)'] = pandas_dataframe_pd['tiempo (programaciondinamica)']

df['n*2^n'] = df.apply(lambda row: row.n * math.pow(2, row.n), axis=1)
df['n*V'] = df.apply(lambda row: row.n * row.V, axis=1)
```

```python
plot = sns.lineplot(x="n", y="tiempo (programaciondinamica)", data=df)
plot = sns.lineplot(x="n", y="n*V", data=df)
plot.legend(["Programacion dinamica (medida)", "n*V"])
plot.set(ylabel="Tiempo");
```

<h1>Heatmap</h1>

```python
df['FB/n*2^n'] = df.apply(lambda row: row['tiempo (fuerzabruta)'] / (row.n * math.pow(2, row.n)), axis=1)
ax1 = sns.lineplot(x="index", y="FB/n*2^n", data=df.reset_index())
plt.ylim(0,0.0005);
plt.ylabel("BU(n,m)/nm")
```

```python
# Graficamos el tiempo de BU en función de n, con series variando m.

legend = []

for x in range(1,21):
    ax1 = sns.lineplot(x="V", y="tiempo (fuerzabruta)", data=df.loc[df['n'] == x]);
    legend.append("n=" + str(x))
    
ax1.legend(legend);
plt.ylabel("FuerzaBruta(n, V))");

```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)

df['FB/n*2^n'] = df['tiempo (fuerzabruta)'] / (df['n'] * (2**df['n']));

ax1 = sns.lineplot(x="n", y="FB/n*2^n", data=df.reset_index(), ax=ax);
plt.ylim(0.0,0.025);

plt.ylabel("BU(n,m)/nm");

plt.setp(ax.lines,linewidth=3)
```
