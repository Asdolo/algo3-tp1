```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
pandas_dataframe_fb = pd.read_csv('output/fuerzabruta_out.csv')
pandas_dataframe_bt = pd.read_csv('output/backtracking_out.csv')
pandas_dataframe_pd = pd.read_csv('output/programaciondinamica_out.csv')

print(pandas_dataframe_bt)
```

```python
# Mergeamos los 3 dataframes en uno solo
df = pandas_dataframe_fb
df['tiempo (backtracking)'] = pandas_dataframe_bt['tiempo (backtracking)']
df['tiempo (programaciondinamica)'] = pandas_dataframe_pd['tiempo (programaciondinamica)']
```

<h1>Fuerza bruta</h1>
<p>Gráfico de comparación entre el tiempo medido y la función O(n*2^n)</p>

```python
a4_dims = (9.0, 4.40)
fig, ax = plt.subplots(figsize=a4_dims)

df['FB/n*2^n'] = df['tiempo (fuerzabruta)'] / (df['n'] * (2**df['n']));

ax1 = sns.lineplot(x="n", y="FB/n*2^n", data=df.reset_index(), ax=ax);
plt.ylim(-0.05,0.05);

plt.ylabel(r"Relación $\frac{FuerzaBruta(n, V)}{n\times n^{2}}$", fontsize=16);
plt.xlabel("n", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
```

El gráfico anterior muestra la relación entre las funciones <b>FuerzaBruta(n,V)</b> y <b>n*2^n</b> .
Notar que la relación tiende a ser constante a medida que aumenta el valor de n.
De esta forma, se puede ver que ambas funciones están relacionadas entre sí, y que únicamente difieren en dicha constante.
La constante calculada en promedio en este caso es:

```python
df['FB/n*2^n'].mean()
```

Lo que quiere decir que la complejidad de FuerzaBruta(n, V) es <b>aproximadamente</b> O(1.4250730697028000 x 10^{-3} n*2^n)

```python
# Graficamos el tiempo de FuerzaBruta en función de n, con series variando V.
a4_dims = (11.7, 5.27)
fig, ax = plt.subplots(figsize=a4_dims)

ax1 = sns.lineplot(x="V", y="tiempo (fuerzabruta)", data=df.loc[df['n'] == 16], label="n=16", ax=ax);
ax1 = sns.lineplot(x="V", y="tiempo (fuerzabruta)", data=df.loc[df['n'] == 17], label="n=17", ax=ax);
ax1 = sns.lineplot(x="V", y="tiempo (fuerzabruta)", data=df.loc[df['n'] == 18], label="n=18", ax=ax);
ax1 = sns.lineplot(x="V", y="tiempo (fuerzabruta)", data=df.loc[df['n'] == 19], label="n=19", ax=ax);
ax1 = sns.lineplot(x="V", y="tiempo (fuerzabruta)", data=df.loc[df['n'] == 20], label="n=20", ax=ax);
ax1 = sns.lineplot(x="V", y="tiempo (fuerzabruta)", data=df.loc[df['n'] == 21], label="n=21", ax=ax);
ax1 = sns.lineplot(x="V", y="tiempo (fuerzabruta)", data=df.loc[df['n'] == 22], label="n=22", ax=ax);
ax1 = sns.lineplot(x="V", y="tiempo (fuerzabruta)", data=df.loc[df['n'] == 23], label="n=23", ax=ax);

#Reverse legends
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(reversed(handles), reversed(labels), title='Líneas', loc='upper right')

plt.ylabel("FuerzaBruta(n,V) [ms]", fontsize=16);
plt.xlabel("V", fontsize=16);

plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)


```

Nótese que, si fijamos el valor de n a un número cualquiera, el tiempo del algoritmo se mantiene aproximadamente constante a medida que aumenta el valor de V.
El gráfico anterior prentende demostrar justamente eso; que la complejidad del algoritmo FuerzaBruta no depende del valor objetivo V, sino que únicamente depende de la longitud del conjunto S.

<h1>Programación dinámica</h1>
<p>Gráfico de comparación entre el tiempo medido y la función O(n*V)</p>

```python
a4_dims = (9.0, 4.8)
fig, ax = plt.subplots(figsize=a4_dims)

df['n*V'] = (df['n'] * df['V']);
df['PD/n*V'] = df['tiempo (programaciondinamica)'] / (df['n'] * df['V']);

ax1 = sns.lineplot(x="n*V", y="PD/n*V", data=df.reset_index(), ax=ax);
plt.ylim(-0.05,0.05);

plt.ylabel(r"Relación $\frac{ProgramaciónDinámica(n, V)}{n\times V}$", fontsize=16);
plt.xlabel(r"$n \times V$", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
```

El gráfico anterior muestra la relación entre las funciones <b>ProgramacionDinámica(n,V)</b> y <b>n*V</b> .
De esta forma, se puede ver que ambas funciones están relacionadas entre sí por una constante.
La constante calculada en promedio en este caso es:

```python
df['PD/n*V'].mean()
```

Lo que quiere decir que la complejidad de ProgramaciónDinámica(n, V) es <b>aproximadamente</b> O(3.084494672339200×10^{-4} n*V)
