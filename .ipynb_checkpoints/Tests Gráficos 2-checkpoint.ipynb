```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
from matplotlib.ticker import MaxNLocator
```

<h1>Test 1</h1>
Fuerza Bruta vs Backtracking (sin podas)

```python
df_test_1_fb = pd.read_csv('output/test-1-fuerzabruta.csv')
del df_test_1_fb['resultado']

df_test_1_bt = pd.read_csv('output/test-1-backtracking.csv')
del df_test_1_bt['resultado']

# Obtengo el promedio en cada tabla
df_test_1_fb = df_test_1_fb.groupby(['n', 'V'], as_index=False, sort=False).mean()
df_test_1_bt = df_test_1_bt.groupby(['n', 'V'], as_index=False, sort=False).mean()

# renombro las columnas en cada dataframe y los mergeo en uno solo
df_test_1_fb.columns = ['n', 'V', 'tiempo (fuerzabruta)']
df_test_1_bt.columns = ['n', 'V', 'tiempo (backtracking)']

df_test_1 = df_test_1_fb
df_test_1['tiempo (backtracking)'] = df_test_1_bt['tiempo (backtracking)']

# grafico
a4_dims = (6.75, 3.975)
fig, ax = plt.subplots(figsize=a4_dims)
sns.set_palette(sns.color_palette("cubehelix", 2))
sns.set_style("whitegrid")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))


ax1 = sns.lineplot(x="n", y="tiempo (fuerzabruta)", data=df_test_1, label='Fuerza Bruta', ax=ax);
ax1 = sns.lineplot(x="n", y="tiempo (backtracking)", data=df_test_1, label='Backtracking', ax=ax);


plt.ylabel("Tiempo de ejecución [ms]", fontsize=16);
plt.xlabel("n", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
```

<h1>Test 2</h1>
Fuerza Bruta vs Backtracking (con podas)

<h3>Poda de medio árbol</h3>
Complejidad practica esperada: O(n*2^n/2)

```python
df_test_2_a_fb = pd.read_csv('output/test-2-a-fuerzabruta.csv')
del df_test_2_a_fb['resultado']

df_test_2_a_bt = pd.read_csv('output/test-2-a-backtracking.csv')
del df_test_2_a_bt['resultado']

# Obtengo el promedio en cada tabla
df_test_2_a_fb = df_test_2_a_fb.groupby(['n', 'V'], as_index=False, sort=False).mean()
df_test_2_a_bt = df_test_2_a_bt.groupby(['n', 'V'], as_index=False, sort=False).mean()

# renombro las columnas en cada dataframe y los mergeo en uno solo
df_test_2_a_fb.columns = ['n', 'V', 'tiempo (fuerzabruta)']
df_test_2_a_bt.columns = ['n', 'V', 'tiempo (backtracking)']

df_test_2_a = df_test_2_a_fb
df_test_2_a['tiempo (backtracking)'] = df_test_2_a_bt['tiempo (backtracking)']

# grafico
a4_dims = (6.75, 3.975)
fig, ax = plt.subplots(figsize=a4_dims)
sns.set_palette(sns.color_palette("cubehelix", 3))
sns.set_style("whitegrid")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))

ax1 = sns.lineplot(x="n", y="tiempo (fuerzabruta)", data=df_test_2_a, label='Fuerza Bruta', ax=ax);
ax1 = sns.lineplot(x="n", y="tiempo (backtracking)", data=df_test_1, label='Backtracking (Sin podas - Experimento 1)', ax=ax);
ax1 = sns.lineplot(x="n", y="tiempo (backtracking)", data=df_test_2_a, label='Backtracking (Poda de medio árbol)', ax=ax);


plt.ylabel("Tiempo de ejecución [ms]", fontsize=16);
plt.xlabel("n", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
```

<h3>Poda de todo el árbol</h3>
Complejidad practica esperada: O(1)

```python
df_test_2_b_fb = pd.read_csv('output/test-2-b-fuerzabruta.csv')
del df_test_2_b_fb['resultado']

df_test_2_b_bt = pd.read_csv('output/test-2-b-backtracking.csv')
del df_test_2_b_bt['resultado']

# Obtengo el promedio en cada tabla
df_test_2_b_fb = df_test_2_b_fb.groupby(['n', 'V'], as_index=False, sort=False).mean()
df_test_2_b_bt = df_test_2_b_bt.groupby(['n', 'V'], as_index=False, sort=False).mean()

# renombro las columnas en cada dataframe y los mergeo en uno solo
df_test_2_b_fb.columns = ['n', 'V', 'tiempo (fuerzabruta)']
df_test_2_b_bt.columns = ['n', 'V', 'tiempo (backtracking)']

df_test_2_b = df_test_2_b_fb
df_test_2_b['tiempo (backtracking)'] = df_test_2_b_bt['tiempo (backtracking)']

# grafico
a4_dims = (6.75, 3.975)
fig, ax = plt.subplots(figsize=a4_dims)
sns.set_palette(sns.color_palette("cubehelix", 2))
sns.set_style("whitegrid")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))

ax1 = sns.lineplot(x="n", y="tiempo (fuerzabruta)", data=df_test_2_b, label='Fuerza Bruta', ax=ax);
ax1 = sns.lineplot(x="n", y="tiempo (backtracking)", data=df_test_2_b, label='Backtracking (Poda de todo el árbol)', ax=ax);

plt.ylim(-0.05,2)

plt.ylabel("Tiempo de ejecución [ms]", fontsize=16);
plt.xlabel("n", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
```

<h3>Poda de todo el arbol menos una rama </h3>
Complejidad practica esperada: O(n²)

```python
df_test_2_c_fb = pd.read_csv('output/test-2-c-fuerzabruta.csv')
del df_test_2_c_fb['resultado']

df_test_2_c_bt = pd.read_csv('output/test-2-c-backtracking.csv')
del df_test_2_c_bt['resultado']

# Obtengo el promedio en cada tabla
df_test_2_c_fb = df_test_2_c_fb.groupby(['n', 'V'], as_index=False, sort=False).mean()
df_test_2_c_bt = df_test_2_c_bt.groupby(['n', 'V'], as_index=False, sort=False).mean()

# renombro las columnas en cada dataframe y los mergeo en uno solo
df_test_2_c_fb.columns = ['n', 'V', 'tiempo (fuerzabruta)']
df_test_2_c_bt.columns = ['n', 'V', 'tiempo (backtracking)']

df_test_2_c = df_test_2_c_fb
df_test_2_c['tiempo (backtracking)'] = df_test_2_c_bt['tiempo (backtracking)']

# grafico
a4_dims = (6.75, 3.975)
fig, ax = plt.subplots(figsize=a4_dims)
sns.set_palette(sns.color_palette("cubehelix", 2))
ax.xaxis.set_major_locator(MaxNLocator(integer=True))

ax1 = sns.lineplot(x="n", y="tiempo (fuerzabruta)", data=df_test_2_c, label='Fuerza Bruta', ax=ax);
ax1 = sns.lineplot(x="n", y="tiempo (backtracking)", data=df_test_2_c, label='Backtracking (Poda de todo el árbol excepto una rama)', ax=ax);

plt.ylim(-0.05,2)


plt.ylabel("Tiempo de ejecución [ms]", fontsize=16);
plt.xlabel("n", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
```
